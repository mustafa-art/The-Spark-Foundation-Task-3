```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline 
import seaborn as sns 
sns.set()
```

```python
data = pd.read_csv('SampleSuperstore.csv')
data.head()
```

```python
data.shape
```

```python
# Check missing Value 
data.isnull().sum()
```

```python
# Check for duplicates 
data.duplicated().sum()
```

- ok in this particular analysis i'll keep the duplicates 

```python
data.head()
```

```python
data['Ship Mode'].value_counts()
```

```python
# i think 'same day ' is not kind a clas so i will remove it but will keep it in the original data 

data_cleaned =data[data['Ship Mode'] !='Same Day']
data_cleaned
```

```python
data_cleaned['Ship Mode'].value_counts()
```

```python
data_cleaned.info()
```

```python
# determine which ship mode has best sales and profits 
avg_sales_by_shipmode = data_cleaned.groupby('Ship Mode')['Sales'].mean()
avg_sales_by_shipmode
```

```python
# plot a bar chart for that 
plt.figure(figsize=(8,6))

avg_sales_by_shipmode.plot(kind='bar')
plt.title('Average sales by each ship mode')
plt.xlabel('Ship Mode')
plt.ylabel('Average Sales')
plt.show()
```

- the Second class has the hieghest sales but not with a big difference 

```python
# let's see is it the same for profit 
avg_profit_by_shipmode = data_cleaned.groupby('Ship Mode')['Profit'].mean()
avg_profit_by_shipmode
```

```python
plt.figure(figsize=(8,6))

avg_profit_by_shipmode.plot(kind='bar')
plt.title('Average profit by Ship mode')
plt.xlabel('Ship Mode')
plt.ylabel('Average Profit')
plt.show()
```

- most of our profits comes from orders shiped in the first class 

```python
data_cleaned['Segment'].value_counts()
```

```python
data_cleaned.groupby('Segment')['Profit'].mean()
```

```python
data_cleaned.groupby(['Segment', 'Category'])['Profit'].mean()
```

```python
# compare the Average sales and profit in one pivot table for segment 
segment_sales_profit = data_cleaned.pivot_table(values=['Sales', 'Profit'], index=['Segment']).sort_values('Profit', ascending=False)
segment_sales_profit
```

```python
# Visualize the results 
segment_sales_profit.plot(kind='bar', figsize=(8,6))
plt.title('Average profit and sales for each segment')
plt.xlabel('Segments')
plt.ylabel('Average sales and Profit')
plt.show()
```

- The 'Home Office' segment has the highest profit and sales, but with a slightly small difference from the 'Corporate' segment, so i guess we find which segment to focus on and can promot

```python
data_cleaned.head()
```

```python
# Let's see in which category particulary we should focus more to gain more profit
segment_category_pivot = data_cleaned.pivot_table(values=['Sales', 'Profit'], index=['Segment', 'Category']).sort_values('Profit', ascending=False)
segment_category_pivot
```

```python
# Viualize the results 
segment_category_pivot.plot(kind='bar', figsize=(8,6))
plt.title('Average Sales and Profit by each Segment for each category')
plt.xlabel('Segment , Category')
plt.ylabel('average Sales and Profit')
plt.show()
```

- The 'home office' segment has the highest sales and profits in technology category, next is 'corporate' segment also with the technology category, we also conclude that the most profitable category is the technology one with all segments by far from other categories 

```python
# ok let's dig deeper and see in which subcategory 
segment_subcateg = data_cleaned.pivot_table(values=['Sales', 'Profit'], index=['Segment', 'Category', 'Sub-Category']).sort_values('Profit', ascending=False)
segment_subcateg
```

```python
segment_subcateg.head(10).plot(kind='bar', figsize=(8,6))
plt.title('the average sales and profits for each sub category')
plt.xlabel('Segment, Category, SubCategory')
plt.ylabel('Average Proft and Sales')
plt.show()
```

- From this we conclude that the most profitable category is the 'Copiers' Forom the 'Technology' Category, also 'Home Office' & 'Corporate' Segments are the most that buy these sub category
